```python
import pandas as pd
```

```python
#creating the dataset
import random
from datetime import datetime,timedelta
```

```python
#parameters
num_entries=1000
start_date=datetime(2024,11,1)
store_ids=["S001","S002","S003","S004","S005","S006"]
product_ids=["P001", "P002", "P003", "P004","P005","P006","P007","P008"]
customer_segments = ["Regular", "Premium", "New"]
```

```python
#generate data
data={
    "Date": [start_date+timedelta(days=random.randint(0,30))for _ in range(num_entries)],
    "Store_id": [random.choice(store_ids)for _ in range(num_entries)],
    "Product_id": [random.choice(product_ids)for _ in range(num_entries)],
    "Units_sold": [random.randint(0,100)for _ in range(num_entries)],
    "Sales Amount": [round(random.uniform(100, 1500), 2) for _ in range(num_entries)],
    "Discount Applied": [round(random.uniform(0, 20), 2) for _ in range(num_entries)],
    "Customer Segment": [random.choice(customer_segments) for _ in range(num_entries)]
}

# Create DataFrame and save to CSV
df = pd.DataFrame(data)
df.to_csv("sales_data.csv", index=False)

print("sales_data.csv created successfully!")
```
